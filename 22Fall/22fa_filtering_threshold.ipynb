```python
import matplotlib.pyplot as plt
import numpy as np
 
file = open("E:/Lab/2022_Spring/data/220511_kallisto/filtering/chr_perc.txt")
perc_str = file.read().splitlines() # list
perc = [float(str) for str in perc_str]
 
plt.figure(figsize=(100,30))
n, bins, patches = plt.hist(x=perc, bins="auto", color='black', alpha=0.5, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xticks(fontsize=80)
plt.yticks(fontsize=80)
plt.xlabel('%Chromatin',fontsize=100)
plt.ylabel('Frequency',fontsize=100)
plt.title('Chromatin Enrichment of All Genes in Mouse TSCs\n',fontsize=100,fontweight='bold')
plt.ylim(ymax=26000)


plt.axvline(x=0.75, ymin=0, ymax=1, linewidth=4, color='r')
plt.text(x=0.75, y=-1100, s='0.75', fontsize=80, color='r', horizontalalignment='center')
 
plt.show()
#plt.savefig('chr%_hist.png')

# threshold = 0.75
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
 
file = open("E:/Lab/2022_Spring/data/220511_kallisto/filtering/total_tpm.txt")
d = file.read().splitlines() # list
e = [math.log(float(z)+0.001,2) for z in d]
 
plt.figure(figsize=(100,30))
n, bins, patches = plt.hist(x=e, bins="auto", color='black', alpha=0.5, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xticks(fontsize=80)
plt.yticks(fontsize=80)
plt.xlabel('log2(TPM)',fontsize=100)
plt.ylabel('Frequency',fontsize=100)
plt.title('Expression Levels of All Genes in Mouse TSCs\n',fontsize=100,fontweight='bold')
maxfreq = n.max()
plt.ylim(ymax=2600)
plt.xlim(xmax=10)
 
plt.axvline(x=-2, ymin=0, ymax=1, linewidth=4, color='r')
plt.text(x=-2, y=-110, s='2', fontsize=80, color='r', horizontalalignment='center')
 
plt.show()
#plt.savefig('total_tpm_hist.png')

# threshold = 2^(-2) = 0.25
```
